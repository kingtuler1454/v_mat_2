Задание 1 - решить систему AX = B

```python
# Создать произвольную нижнетреугольную матрицу А 5 порядка 
# (не унитреугольную), вектор B произвольный. Решить систему AX = B.

# Создание нижнетреугольной матрицы A (не унитреугольной)
A = np.tril(np.random.randint(1, 10, size=(5, 5)))

# Создание вектора B
B = np.random.randint(1, 10, size=(5,))

# Нахождение обратной матрицы A^(-1)
A_inv = np.linalg.inv(A)
# Решение 
X_1 = np.dot(A_inv, B)

# Проверка решения
X_2 = np.linalg.solve(A, B)

print("Матрица A:\n", A)
print("Вектор B:\n", B)
print("\nРешение с помощью обратной матрицы:\n", X_1)
print("Решение с помощью np.linalg.solve:\n", X_2)
```

Задание 2 - Решить систему, используя LU разложение

```python
# функция разложения из 1-ой части лабораторной
def LU_decomposition(matrix):
    n = matrix.shape[0]
    L = np.eye(n)
    U = np.copy(matrix)

    for k in range(n-1):
        for i in range(k+1, n):
            L[i, k] = U[i, k] / U[k, k]
            for j in range(k, n):
                U[i, j] -= L[i, k] * U[k, j]
                
    return L, U


# Алгоритм:
# Находим L и U =>  
# L*U * X = B
# Пусть Y = U*X,  тогда L*Y=B,  и находим  Y = (L^-1)*B
# Затем находим X = (U^-1)*Y

# Задание системы уравнений
matrix_A = np.array([[14.4, -5.3, 14.3, -12.7],
                   [23.4, -14.2, -5.4, 2.1],
                   [6.3, -13.2, -6.5, 14.3],
                   [5.6, 8.8, -6.7, -23.8]
                   ], dtype=float)

vector_B = np.array([[14.4], [6.6], [9.4], [7.3]])

print("Левая часть(матрица):\n",matrix_A, "\nПравая часть(вектор)\n",vector_B,"\nдолжно получиться: х1~=-1,73; х2~=-3,5; х3~=-0,266; х4~=-1,93")



L, U = LU_decomposition(matrix_A)

# нахождение обратных матриц
U_inverse = np.linalg.inv(U)
L_inverse = np.linalg.inv(L)

# решение Y = (L^-1)*B
Y = L_inverse.dot(vector_B)

# решение X = (U^-1)*Y
X_lu = U_inverse.dot(Y)

print("\nМатрица L:\n", L)
print("\nМатрица U:\n", U)

print("\nРезультат:\n", X_lu)
```

Задание 3 - ту же систему, используя QR разложение методом Хаусхолдера

```python
# метод Хаусхолдера (отражений) из 1 части лабораторной 
def householder(matrix):
    (r, c) = np.shape(matrix)
    Q = np.identity(r)
    R = np.copy(matrix)

    for i in range(r - 1):
        x = R[i:, i]
        e = np.zeros_like(x)
        e[0] = np.linalg.norm(x)
        u = x - e
        v = u / np.linalg.norm(u)
        Q_i = np.identity(r)
        Q_i[i:, i:] -= 2.0 * np.outer(v, v)
        R = np.dot(Q_i, R)
        Q = np.dot(Q, Q_i)

    return Q, R


# Алгоритм:
# Находим Q и R =>  
# Q*R * X = B
# Пусть Y = R*X,  тогда Q*Y=B,  и находим  Y = (Q^-1)*B
# Затем находим X = (R^-1)*Y

print("Левая часть(матрица):\n", matrix_A, "\nПравая часть(вектор)\n", vector_B)


# находим QR разложение
Q, R = householder(matrix_A)

# нахождение обратных матриц
R_inverse = np.linalg.inv(R)
Q_inverse = np.linalg.inv(Q)

# решение Y = (Q^-1)*B
Y = Q_inverse.dot(vector_B)

# решение X = (R^-1)*Y
X_qr = R_inverse.dot(Y)

print("\nМатрица Q:\n", Q)
print("\nМатрица U:\n", R)

print("\nРезультат:\n", X_qr)
```

Задание 4 - Метод Зейделя (точность до 10^-3)

```python
from prettytable import PrettyTable

# х1 = 1.575  х2 = 1.5724  х3 = 1.57697

# Задание системы уравнений
A = np.array([[2.15, -1.18, 0.57],[1.24, 0.62, -0.95],
              
              [1.72, -0.83, 1.57]])
B = np.array([2.43,1.43,  3.88])

# Задание начальные приближения для решения системы
x_prev = np.zeros(3)
x_curr = np.zeros(3)

# Задание точности
epsilon = 1e-3

# Инициализация таблицы
table = PrettyTable()
table.field_names = ["№ итерации", "x_1", "x_2", "x_3"]

# Итерационный процесс
iteration = 1
while True:
    # вычисление нового значения x_curr для каждого 
    # уравнения системы по ф-ле метода Зейделя
    for i in range(len(A)):
        x_curr[i] = (B[i] - np.dot(A[i, :i], x_curr[:i]) - np.dot(A[i, i+1:], x_prev[i+1:])) / A[i, i]
    
    # Проверка достижения заданной точности
    if np.all(np.abs(x_curr - x_prev) < epsilon):
        break
    
    # Обновление предыдущих значений
    x_prev = x_curr.copy()
    
    # Запись текущих значений в таблицу
    table.add_row([iteration, round(x_curr[0], 4), round(x_curr[1], 4), round(x_curr[2], 4)])
    
    iteration += 1

# Вывод таблицы
print(table)
```

Задание 5 - Найти псевдорешение системы

```python
# X = ( A_transp * A )^-1 * A_transp * B

# Задание системы уравнений
matrix_A_5 = np.array([[14.4, -5.3, 14.3],
                   [23.4, -14.2, -5.4],
                   [6.3, -13.2, -6.5],
                   [5.6, 8.8, -6.7]
                   ], dtype=float)

vector_B_5 = np.array([[14.4], [6.6], [9.4], [7.3]])

print("Левая часть(матрица):\n", matrix_A_5, "\nПравая часть(вектор)\n", vector_B_5)

# транспонируем матрицу
matrix_A_5_transp = matrix_A_5.transpose()

# A_transp * A 
tmp_1 = np.dot(matrix_A_5_transp, matrix_A_5)

# ( A_transp * A )^-1
tmp_2 = np.linalg.inv(tmp_1)

# ( A_transp * A )^-1 * A_transp * B
tmp_3 = np.dot(np.dot(tmp_2, matrix_A_5_transp), vector_B_5)

print("\nПсевдорешение:\n", tmp_3)
```
