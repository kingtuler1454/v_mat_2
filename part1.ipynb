Создать   две   матрицы   из   случайных   вещественных   чиселподходящего   размера.
Найти   их   произведение   тремя   способами:
используя   векторный   алгоритм   умножения   матриц;
используя матричный алгоритм, записав матрицу С по столбцам;
 проверив спомощью функции np.dot

```python
# Создаем две случайные матрицы подходящего размера, например, 3x4 и 4x2.
matrix1 = np.random.rand(3, 4)
matrix2 = np.random.rand(4, 2)

# Выводим исходные матрицы
print("Исходная матрица 1:")
print(matrix1)
print("Исходная матрица 2:")
print(matrix2)

# 1. Используем векторный алгоритм умножения матриц
result1 = np.dot(matrix1, matrix2)

# 2. Используем матричный алгоритм, записывая матрицу C по столбцам
result2 = np.zeros((matrix1.shape[0], matrix2.shape[1]))
for i in range(matrix2.shape[1]):
    result2[:, i] = np.dot(matrix1, matrix2[:, i])

# 3. Используем функцию np.dot
result3 = np.dot(matrix1, matrix2)

# Выводим результаты
print("Результат с использованием векторного алгоритма:")
print(result1)
print("Результат с использованием матричного алгоритма, записанный по столбцам:")
print(result2)
print("Результат с использованием функции np.dot:")
print(result3)
```

Создать вектор-строку 1x8 из случайных целых чисел. Вы-числить   норму  ‖x‖3  самостоятельно   написанной   функцией   ипроверить результат с помощью linalg.norm()

```python
# Создаем случайную вектор-строку 1x8 с целыми числами
x = np.random.randint(1, 10, 8)

# Выводим исходный вектор
print("Исходный вектор:")
print(x)

# Самостоятельно написанная функция для вычисления L3-нормы
def custom_l3_norm(vector):
    norm = np.sum(np.abs(vector) ** 3) ** (1/3)
    return norm

# Вычисляем L3-норму с помощью самостоятельно написанной функции
custom_norm_result = custom_l3_norm(x)

# Вычисляем L3-норму с использованием linalg.norm()
numpy_norm_result = np.linalg.norm(x, ord=3)

# Выводим результаты
print("L3-норма, вычисленная самостоятельно:", custom_norm_result)
print("L3-норма, вычисленная с использованием linalg.norm():", numpy_norm_result)
```

Создать матрицу из случайных целых чисел. Найти спек-тральную норму матрицы с помощью самостоятельно написанногоалгоритма   (можно   использовать   функцию   для   нахождения   соб-ственных значений), проверить результат с помощью linalg.norm()

```python
# Создаем матрицу случайных целых чисел, например, 3x3
matrix = np.random.randint(-10, 10, (3, 3))

# Выводим исходную матрицу
print("Исходная матрица:")
print(matrix)

# Самостоятельно написанный алгоритм для вычисления спектральной нормы
def custom_spectral_norm(matrix):
    eigenvalues = np.linalg.eigvals(matrix)  # Находим собственные значения
    spectral_norm = np.max(np.abs(eigenvalues))  # Максимальное собственное значение по модулю
    return spectral_norm

# Вычисляем спектральную норму с помощью самостоятельно написанного алгоритма
custom_norm_result = custom_spectral_norm(matrix)

# Вычисляем спектральную норму с использованием linalg.norm()
numpy_norm_result = np.linalg.norm(matrix, ord=2)

# Выводим результаты
print("Спектральная норма, вычисленная самостоятельно:", custom_norm_result)
print("Спектральная норма, вычисленная с использованием linalg.norm():", numpy_norm_result)
```

Для вектора, созданного в третьем пункте, найти отраже-ние   Хаусхолдера,   которое   обнуляет   его   координаты   с   4   по   8.Например,     вектор     (1,2,3,4,5,6,7,8)     переводит     в     вектор(1,2,c,0,0,0,0,0)

```python
# Создаем исходный вектор v=x−∣∣x∣∣⋅e k

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])

# Выбираем диапазон координат, которые нужно обнулить (с 4 по 8)
k = 4
v = x.copy()
v[k-1:] = 0

# Вычисляем ||x||
norm_x = np.linalg.norm(x)

# Вычисляем вектор отражения
v = v - norm_x * np.eye(len(x))[:, k-1]

# Выводим вектор отражения
print("Вектор отражения Хаусхолдера:")
print(v)
```

Для матрицы, созданной в 4 пункте, найти LU разложениесамостоятельно   написанной   функцией   и   проверить   результат   спомощью перемножения матриц.

```python
# Функция для вычисления LU-разложения
def lu_decomposition(matrix):
    n = matrix.shape[0]
    L = np.eye(n)  # Создаем начальную матрицу L как единичную
    U = matrix.copy()  # Создаем копию исходной матрицы для U

    for key in range(n):
        for i in range(key+1, n):
            if U[key, key] == 0:
                raise ValueError("LU-разложение невозможно: ведущий элемент равен нулю.")
            factor = U[i, key] / U[key, key]
            L[i, key] = factor
            U[i, key:] -= factor * U[key, key:]

    return L, U

# Создаем матрицу (предполагая, что у вас уже есть созданная матрица)
matrix = np.array([[2, 3, 1], [4, 7, 7], [1, 5, 5]], dtype=float)

# Вычисляем LU-разложение
L, U = lu_decomposition(matrix)

# Проверяем результат, перемножив матрицы L и U
result = np.dot(L, U)

# Выводим результаты
print("Матрица L:")
print(L)
print("Матрица U:")
print(U)
print("Результат перемножения L и U:")
print(result)
```

Для матрицы, созданной в 4 пункте, найти QR разложениевсеми рассмотренными методами, проверить результат с помощьюперемножения матриц и с помощью функции np.linalg.qr.

```python
# Функция для вычисления QR-разложения с помощью метода Грама-Шмидта
def gram_schmidt_qr(matrix):
    n, m = matrix.shape
    Q = np.zeros((n, m))
    R = np.zeros((m, m))

    for j in range(m):
        v = matrix[:, j]
        for i in range(j):
            R[i, j] = np.dot(Q[:, i], matrix[:, j])
            v = v - R[i, j] * Q[:, i]
        R[j, j] = np.linalg.norm(v)
        Q[:, j] = v / R[j, j]

    return Q, R



# Вычисляем QR-разложение с помощью метода Грама-Шмидта
Q_gs, R_gs = gram_schmidt_qr(matrix)

# Вычисляем QR-разложение с использованием функции np.linalg.qr
Q_np, R_np = np.linalg.qr(matrix)

# Проверяем результаты с помощью перемножения матриц и сравнения с библиотечной функцией
result_gs = np.dot(Q_gs, R_gs)
result_np = np.dot(Q_np, R_np)

# Выводим результаты
print("Матрица Q (метод Грама-Шмидта):")
print(Q_gs)
print("Матрица R (метод Грама-Шмидта):")
print(R_gs)
print("Результат перемножения Q и R (метод Грама-Шмидта):")
print(result_gs)
print("Матрица Q (с использованием np.linalg.qr):")
print(Q_np)
print("Матрица R (с использованием np.linalg.qr):")
print(R_np)
print("Результат перемножения Q и R (с использованием np.linalg.qr):")
print(result_np)
```
